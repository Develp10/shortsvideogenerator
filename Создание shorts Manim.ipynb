```python
from manim import *

class CodeFromString(Scene, Camera):
    
    def construct(self):

        image = ImageMobject("fluid-gradient_cr.png").scale(2)
        self.add(image)
        
        self.add_sound("output.mp3", gain=-20)  # убавляем звук
        self.add_sound("voice audio2.mp3")

        code = '''
            # pip install elevenlabs
            from elevenlabs import generate, play, save
            
            text = 'Это текст, который нужно озвучить'
            audio = generate(
            	text = text,  
            	voice = 'Bella', 
            	model = 'eleven_multilingual_v2',
            	)
            
            play(audio)
            # save(audio, 'voice.mp3')
            '''
        rendered_code = Code(code=code, 
                             tab_width=4, 
                             background="window",  # "rectangle"
                             language="Python", 
                             font="Monospace",
                             insert_line_no=False, 
                             line_spacing=1, 
                             font_size = 33,
                             style=Code.styles_list[7])   # random.choice
        
        self.play(Write(rendered_code[:2]))
        self.wait(1)
        
        for line in rendered_code[-1]:
            self.play(Write(line))
            self.wait(1)

        self.play(FadeOut(rendered_code))
        self.add_sound("subscribe.mp3")
        self.wait(1)
        text = Text("@data_analysis_ml", color=BLACK).scale(2)
        self.play(Write(text))
        self.wait(1)



%manim -v WARNING -r 1080,2160 CodeFromString
# %manim -v WARNING -r 200,400 CodeFromString
```
